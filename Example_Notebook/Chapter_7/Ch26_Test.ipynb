###  連續頁面查找  
- 函示化
- 查找最大頁數

```python
def get_car_data(url):
    datas=[]
    chrome=None
    try:
        chrome=get_chrome(url)
        soup=BeautifulSoup(chrome.page_source,'lxml')
        cars=soup.find(id="result-items").find_all('a')
        # 查找該頁的車輛資訊
        for i,car in enumerate(cars):
            try:    
                title=car.find('span',class_="ib-it-text").text.strip()
                link=url+car.get('href')
                image_url=car.find('img').get('data-src')
                location=car.find('span',class_="ib-ii-item").text.strip()
                price=car.find('span',class_="ib-price").text.split()[0]
                years,km=[span.text.strip() for span in car.find_all('span',class_="ib-icon")]           
                #print(f'No.{i+1} {title}',link,image_url,location,price,years,km,sep='\n')
                datas.append([title,link,image_url,location,price,years,km])
                #print('---------------------------------------------------')
            except Exception as e:
                print(e)           
    
    except Exception as e:
        print(e)
    finally:
        # 進行關閉
        if chrome is not None:
            chrome.quit()
            
    return datas
```

```python
datas=get_car_data(url)
len(datas)
```

### 查找最大頁數

```python
pages=eval(soup.find('div',class_="_container_ivprf_1").find_all('button')[-2].text)
pages
```

### 組合url  
- https://auto.8891.com.tw/?page=1

```python
datas=[]
for i in range(pages):
    url=f'https://auto.8891.com.tw/?page={i+1}'
    print(url)
    data=get_car_data(url)
    # 確定該頁數有資料
    if data!=[]:
        # 使用extend合併資料
        datas.extend(data)
    
datas
```

```python
len(datas)
```

```python
df=pd.DataFrame(datas,columns=['title','link','image_url','location','price','years','km'])
df
```
