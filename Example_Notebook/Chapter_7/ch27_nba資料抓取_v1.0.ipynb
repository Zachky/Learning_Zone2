## 函式化

```python
def get_nba_data(date):
    print(date)
    api_url='https://tw.global.nba.com/stats2/scores/daily.json'
    data={

        'countryCode': 'TW',
        'gameDate': date,
        'locale': 'zh_TW',
        'tz': '+8'
    }
    
    resp=requests.get(api_url,params=data)
    if resp.status_code==200:
        datas=json.loads(resp.text)
        all_data=[]
        if datas['payload']['date'] is not None:            
            games=datas['payload']['date']['games']
            teams=['awayTeam','homeTeam']
           
            for game in games:          
                for team in teams:
                    team=game[team]
                    name=team['profile']['name']
                    temp_data=[date,name]
                    score=team['score']
                    score=[score['q1Score'],score['q2Score'],score['q3Score'],score['q4Score'],score['score']]
                    print(name,score)
                    temp_data+=score
                    all_data.append(temp_data)

                print('---------------------------------------------------------------------------------')
        else:
            return None
            
    return all_data
```

## 連續資料抓取

```python
import time

start=dt.datetime(2019,1,1)
end=dt.datetime(2019,12,31)

datas=[]
while start<end:
    #time.sleep(1)
    start+=dt.timedelta(1)
    date=start.strftime('%Y-%m-%d')    
    data=get_nba_data(date)   
    
    if data is None:
        print('當天無比賽')
        continue
        
    datas.extend(data)
    
datas
    
```

## 進行分數加總  
使用zip

```python
x=[1,2,3,4]
y=[10,20,30,40]

list(zip(x,y))

for x in list(zip(x,y)):
    print(sum(x))
```

```python
datas=[]
for i in range(len(round_score.values)):
    if i%2==1:
        datas.append(list(zip(round_score.values[i-1],round_score.values[i])))
        
datas
    
```

```python
df4=pd.DataFrame(datas,columns=['round1','round2','round3','round4'])
df4
```

```python
sum(df4.iloc[0,0])
```

## 每行計算總分

```python
for col in df4:
    df4[col]=df4[col].apply(lambda x:sum(x))    
df4     
```

```python
df4['round1']%2==1
```

```python
len(df4[df4['round1']%2==1])/len(df4)
```

## 偶數機率

```python
print('偶數機率')
for col in df4:
    print('{}:{:.2%}'.format(col,len(df4[df4[col]%2==0])/len(df4))) 
```

## 奇數機率

```python
print('奇數機率')
for col in df4:
    print('{}:{:.2%}'.format(col,len(df4[df4[col]%2!=0])/len(df4))) 
```
