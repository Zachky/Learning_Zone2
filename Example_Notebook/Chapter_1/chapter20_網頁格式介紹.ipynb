## yahoo é€£çµæŸ¥æ‰¾

```python
import requests
from bs4 import BeautifulSoup

def getSoup(url):
    resp=requests.get(url)
    resp.encoding='utf-8'
    
    if resp.status_code==200:
        return BeautifulSoup(resp.text,'lxml')
    
    return None    
```

```python
url='https://tw.yahoo.com/'
```

```python
soup=getSoup(url)
soup
```

### æŸ¥æ‰¾æ‰€æœ‰é€£çµ

```python
soup.find_all('a')
```

```python
for link in soup.find_all('a'):
    print(link.text)
    print(link.get('href'))
```

```python
datas=[[link.text,link.get('href')] for link in soup.find_all('a') if link.text!='']
datas
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(datas,columns=['åç¨±','é€£çµ'])
df
```
