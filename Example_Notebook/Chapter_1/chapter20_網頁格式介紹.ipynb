## yahoo 連結查找

```python
import requests
from bs4 import BeautifulSoup

def getSoup(url):
    resp=requests.get(url)
    resp.encoding='utf-8'
    
    if resp.status_code==200:
        return BeautifulSoup(resp.text,'lxml')
    
    return None    
```

```python
url='https://tw.yahoo.com/'
```

```python
soup=getSoup(url)
soup
```

### 查找所有連結

```python
soup.find_all('a')
```

```python
for link in soup.find_all('a'):
    print(link.text)
    print(link.get('href'))
```

```python
datas=[[link.text,link.get('href')] for link in soup.find_all('a') if link.text!='']
datas
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(datas,columns=['名稱','連結'])
df
```

```python
df.to_csv('yahoo_link.csv',encoding='utf-8-sig',index=0)
```
