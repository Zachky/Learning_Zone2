## 撰寫轉換函示

```python
def convert_date(date):
    date=str(date)
    year=date[:4]
    math=date[4:6]
    day=date[6:8]
    hh=date[8:10]
    mm=date[10:12]
    ss=date[12:14]
    
    return f'{year}/{math}/{day} {hh}:{mm}:{ss}'   
   
```

```python
convert_date('20210219105945')
```

```python
df2['mday'].apply(convert_date)
```

```python
df2['mday']=df2['mday'].apply(convert_date)
df2.head()
```

## 建立資料庫  
mydb  
CREATE TABLE "youbike_data" (
	"id"	INTEGER,
	"date"	TEXT,
	"sno"	BLOB,
	"sna"	TEXT,
	"tot"	INTEGER,
	"sbi"	INTEGER,
	"bemp"	INTEGER,
	"act"	INTEGER,
	PRIMARY KEY("id" AUTOINCREMENT)
)


```python
count=0

try:
    conn=sqlite3.connect('mydb.db')
    print('open db.')
    curosr=conn.cursor()    
   
    
        
    insert_str='insert into youbike_data(date,sno,sna,tot,sbi,bemp,act) values("{}","{}","{}",{},{},{},{});'
       
    
    for data in datas:
        #檢查是否有重複資料
        select_str=f'select date,sno from youbike_data where date="{data[0]}" and sno="{data[1]}"'
        
        if len(list(curosr.execute(select_str)))==0:        
            count+=1
            curosr.execute(insert_str.format(data[0],data[1],data[2],data[3],data[4],data[5],data[6]))
                       
    conn.commit()
    
    print('寫入資料庫成功!')
    print(f'更新{count}筆資料')
        
except Exception as e:
    print(e)
finally:
    conn.close()
    print('close db.')
```

## 使用scheduler套件  
可排程同時運行

```python
import time
import sched
import datetime


#生成scheduler物件
s = sched.scheduler(time.time, time.sleep) 

def print_time(count):   
    print(count,datetime.datetime.now())


count=0
s.enter(1, 0, print_time, (count,)) 
count+=1
s.enter(3, 1, print_time, (count,))

s.run()
```
