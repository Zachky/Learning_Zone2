### 函式增加headers指定參數

```python
import requests

def getSoup(url):
    resp=requests.get(url,headers=headers)
    if resp.status_code==200:
        return BeautifulSoup(resp.text,'lxml')
    
    return None
```

## 排序價格  
將價格轉換成float

```python
df2['price'].values
```

```python
import re
#使用正則式
def getRealPrice(price):
    result=re.search(r'\d+.*\d+',price)
    if result!=None:
        return result.group()
    else:
        return 99999
    
```

```python
getRealPrice('158.7萬')
```

```python
df2['price']=list(map(getRealPrice,df2['price'].values))
df2['price'].values
```

```python
df2['price']=df2['price'].astype(float)
df2['price']
```

## 整合成全頁數抓取

```python
url='https://auto.8891.com.tw'
car='altis\n'
page_url='https://auto.8891.com.tw/?key=altis&page={}'


chrome=tools.getChrome(url)

if chrome!=None:
    print(f'查詢{car}')
    xpath='/html/body/div[5]/div/div[3]/div[1]/form/input[1]'
    element=chrome.find_element_by_xpath(xpath)
    element.clear()
    element.send_keys(car)

    time.sleep(3)
    
    print('查詢資料中...')    
    soup=BeautifulSoup(chrome.page_source,'lxml')
    pages=int(soup.find(id="pagination-all").text.strip())    
    chrome.quit()   
    
    print(f'最大頁次:{pages}')
    
    all_car_list=[]
    for page in range(1,pages):
        print(page_url.format(page))    
        chrome=tools.getChrome(page_url,hide=True)
        
        if chrome!=None:
            time.sleep(2)        
    
            soup=BeautifulSoup(chrome.page_source,'lxml')
            datas=soup.find(id="search-result").find_all('a')
            car_list=[[data.find('span',class_="ib-it-text").text.strip(),
              data.find('span',class_="ib-price").text.strip(),
              url+data.get('href')] for data in datas if data.find('span',class_="ib-it-text")!=None]

   
            all_car_list.extend(car_list)
    
        chrome.quit()
        time.sleep(1)

all_car_list
```
