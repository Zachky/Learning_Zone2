# 臺灣觀光年曆 https://www.taiwan.net.tw/m1.aspx?sNo=0001019   
1.使用get傳遞參數  
2.連續頁面資料抓取


```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.taiwan.net.tw/m1.aspx?sNo=0001019&keyString=^^^^20200701^20201231'

main_url='https://www.taiwan.net.tw/m1.aspx'

```

```python
form_data={
    "sNo":"0001019",
    "keyString":"^^^^20200701^20201231",
    "page":1
}
```

```python
resp=requests.get(main_url)
resp
```

# 使用get傳遞參數 

```python
resp=requests.get(main_url,params=form_data)

resp
```

```python
if resp.status_code==200:
    soup=BeautifulSoup(resp.text,'lxml')
    #print(soup)
    
    lis=soup.find('ul',class_='columnBlock-list').find_all('li')
    
    print(len(lis))
    
    for li in lis:
        print(li.find('a')['href'])
        print(li.text.strip())
```

# 找出總頁數 

```python
pages=soup.find('div',class_='title news-title').find_all('span')[2]

pages=pages.b.text

pages
```

```python
for page in range(1,int(pages)+1):
    from_data={
        "sNo":"0001019",
        "keyString":"^^^^20200701^20201231",
        "page":str(page)
        
    }
    
    print('分析第%d頁，按任一鍵開始'%page)
    
    resp=requests.get(main_url,params=from_data)
    
    if resp.status_code==200:
        soup=soup=BeautifulSoup(resp.text,'lxml')
        #print(soup)

        lis=soup.find('ul',class_='columnBlock-list').find_all('li')

        print('筆數:%d'%len(lis))

        for li in lis:
            print(li.find('a')['href'])
            print(li.text.strip()) 
```
