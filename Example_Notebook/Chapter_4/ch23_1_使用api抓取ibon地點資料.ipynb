```python
import requests
from bs4 import BeautifulSoup
```

```python
def getSoup(url):
    resp=requests.get(url)
    resp.encoding='utf-8'
    if resp.status_code==200:
        soup=BeautifulSoup(resp.text,'lxml')
        return soup
    
    return None
```

```python
url='http://www.ibon.com.tw/retail_inquiry.aspx#gsc.tab=0'
```

## 整理資料

```python
#傳送的表單
formData={
    'strTargetField': 'COUNTY',
    'strKeyWords':'台北市'
}

resp=requests.post(request_url,data=formData)              
resp.text
```

## 組成函示

```python
def getIbonData(city):  
    datas=[]
    formData={
        'strTargetField': 'COUNTY',
        'strKeyWords':city
    }
    
    resp=requests.post(request_url,formData)    
    if resp.status_code==200:
        soup=BeautifulSoup(resp.text,'lxml')   
        trs=soup.find('table').find_all('tr')        
        for tr in trs: 
            tds=tr.find_all('td')
            datas.append([td.text.strip() for td in tds])       
    else:
        print(city+"抓取錯誤!")
    
    
    return datas  
```

## 程式練習　　
請排序分店數，轉成二維串列或dict進行降序

```python
ibon_data={}
for sheet in sheets:
    print(sheet)
    data=pd.read_excel('ibon便利生活站地點.xlsx',sheet_name=sheet,index_col=0)
    print(f'共{len(data)}各分店')
    ibon_data[sheet]=len(data)
    
```

```python
datas=sorted(ibon_data.items(),key=lambda k:k[1],reverse=True)
datas
```
