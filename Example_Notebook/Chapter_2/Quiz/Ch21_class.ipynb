# 查找表格

```python
main_url='https://rbdkd67zcc92ou2hpwdvxa-on.drv.tw/www/test_table1.html'
```

```python
soup=tools.getSoup(main_url)
soup
```

```python
trs=soup.find(id="Table").find_all('tr')
trs
```

```python
trs[0]
```

```python
for th in trs[0].find_all('th'):
    print(th.text)
```

## 輸出二維串列

```python
datas=[]
for i,tr in enumerate(trs[:-1]):
    data=[]
    if i==0:
        for th in tr.find_all('th'):
            print(th.text.strip()) 
            data.append(th.text.strip())
        
    else:       
        for td in tr.find_all('td'):
            print(td.text.strip())    
            data.append(td.text.strip())
            
    datas.append(data)
    print("==================================================")
    
datas
```

```python
pd.DataFrame(datas).to_csv('人事行政局資料.csv',encoding='utf-8-sig')
```

## 查找國泰世華匯率

```python
url='https://accessibility.cathaybk.com.tw/exchange-rate-search.aspx'
```

```python
soup=tools.getSoup(url)
soup
```

```python
trs=soup.find('table').find_all('tr')
trs
```

```python
datas=[]
for i,tr in enumerate(trs):    
    if i==0:
        datas.append([th.text.strip() for th in tr.find_all('th')])       
    else:       
        datas.append([td.text.strip() for td in tr.find_all('td')])               
    
datas
```

```python
pd.DataFrame(datas[1:],columns=datas[0]).to_csv('國泰世華匯率.csv',encoding='utf-8-sig')
```
