## 爬找熱門新聞

```python
import requests
from bs4 import BeautifulSoup

def get_soup(url):
    try:
        resp=requests.get(url)      
        if resp.status_code==200:
            resp.encoding='utf-8'
            
            return BeautifulSoup(resp.text,'lxml')
        
    except Exception as e:
        print(e)    
    
    return None
        
```

```python
url='https://tw.news.yahoo.com/most-popular'
```

```python
soup=get_soup(url)
soup
```

### 指定新聞區塊

```python
data=soup.find('div',class_="Cf")
data
```

## 查找所有新聞
- find_all

```python
datas=soup.find_all('div',class_="Cf")
datas
```

```python
count=0
news_list=[]
for data in datas:
    title=data.find('a').text.strip()
    news_url='https://tw.news.yahoo.com'+data.find('a').get('href')
    image_url=data.find('img').get('src')
    
    print(f'{count+1}.{title}')
    print(news_url)
    print(image_url)
    
    news_list.append([title,news_url,image_url])
    count+=1   
    
news_list   
```

### 推導式寫法

```python
news_list=[[data.find('a').text.strip(),'https://tw.news.yahoo.com'+data.find('a').get('href'),
 data.find('img').get('src')] for data in soup.find_all('div',class_="Cf") ]
news_list
```

### 儲存csv格式

```python
import pandas as pd
```

```python
df=pd.DataFrame(news_list,columns=['title','url','image'])
df
```

```python
df.to_csv('yahoo-most-popular.csv',encoding='utf-8-sig')
```

### 查找所有新聞

```python
datas=[]
for dt in dts:
    title=dt.find('a').find('h3').text.strip()
    info=dt.find('a').find('p').text.strip()
    link=domain_url+dt.find('a').get('href')
    
    image_url=dt.find('a').find('img').get('data-src')
    image_url=image_url[image_url.index('?u=')+len('?u='):image_url.index('&x')]
        
    print(title,info,link,image_url)
    datas.append([title,info,link,image_url])
    
datas
```

```python
df=pd.DataFrame(datas,columns=['title','info','link','image_url'])
df
```

```python
df.to_csv('nba_udn.csv',encoding='utf-8-sig')
```

### 表格爬找  
- table
- thead<tr><th>
- tbody<tr><td>

```python
url='https://www.cathaybk.com.tw/cathaybk/personal/deposit-exchange/rate/currency-billboard/'
```

```python
soup=get_soup(url)
soup
```

```python
table=soup.find('table',class_="cubinvest-table_rate ratepc")
table
```

### tobody 

```python
trs=table.find('tbody').find_all('tr')
trs
```

```python
tr=trs[0]
tr.find_all('td')
```

```python
for td in tr.find_all('td'):
    print(td.text.strip())
```

### 查找所有tr

```python
datas=[]
for tr in trs:
    data=[]
    for td in tr.find_all('td'):
        print(td.text.strip(),end='\t')
        data.append(td.text.strip())
    print(data)
    datas.append(data)

datas
```

### 推導式

```python
datas=[]
for tr in trs:
    data=[td.text.strip() for td in tr.find_all('td')]   
    datas.append(data)

datas
```

```python
datas
```

### thead

```python
thead=table.find('thead')
thead
```

```python
columns=[]
for tr in thead.find_all('tr',class_="m_display"):
    for td in tr.find_all('th'):
        print(td.text.strip(),end='\t')
        columns.append(td.text.strip())
    print()

columns
```

```python
columns.insert(0,'幣別')
columns
```

### 儲存csv

```python
pd.DataFrame(datas,columns=columns).to_csv('國泰世華匯率.csv',encoding='utf-8-sig')
```
