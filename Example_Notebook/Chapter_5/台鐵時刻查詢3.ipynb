```python
import requests
from bs4 import BeautifulSoup
```

```python
def getSoup(url,post_data=None,get_data=None):
    
    if post_data!=None:
        resp=requests.post(url,post_data)
    elif get_data!=None:
        resp=requests.get(url,get_data)
    else:    
        resp=requests.get(url)
        
    resp.encoding='utf-8'    
    if resp.status_code==200:
        soup=BeautifulSoup(resp.text,'lxml')
        return soup
    
    return None
```

## 函式化

```python
def getTicks(start_station,end_station,date,start_time,end_time='23:59'):
    datas=[]
    print(f'{start_station}==>{end_station},{date}')
    api_url='https://www.railway.gov.tw/tra-tip-web/tip/tip001/tip112/querybytime'
    #查找csrf code
    soup=getSoup(url)
    csrf_code=soup.find(id="queryForm").find('input').get('value')
    
    form_data={
        '_csrf': csrf_code,
        'trainTypeList': 'ALL',
        'transfer': 'ONE',
        'startStation': start_station,
        'endStation': end_station,
        'rideDate': date,       
        'startOrEndTime': 'true',
        'startTime': start_time,
        'endTime': end_time
    }
    
    #print(form_data)
    
    soup=getSoup(api_url,post_data=form_data)
    trs=soup.find('table').find('tbody').find_all('tr',class_="trip-column")
    data=[]
    for th in soup.find('table').find('tbody').find('tr').find_all('th'):
        #print(th.text.strip(),end='\t')
        data.append(th.text.strip())
    datas.append(data)
    print()
   
    for tr in trs:  
        data=[]
        for i,td in enumerate(tr.find_all('td')):
            if i==0:
                #print(td.find('a').text.strip(),end='\t')
                data.append(td.find('a').text.strip())
            else:
                #print(td.text.strip(),end='\t')
                data.append(td.text.strip())
            
        #print()  
        datas.append(data)

    
    return datas    
```
