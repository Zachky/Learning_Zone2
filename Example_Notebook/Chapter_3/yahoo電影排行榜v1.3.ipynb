## 函式化

```python
def get_moviedata(url,show=None):
    soup=get_soup(url)
    main=soup.find('div',class_='movie_intro_info')
    image_url=main.find('img').get('src')
    video_url=main.find('a',class_='btn_s_vedio').get('href')
    imdb=main.find('div',class_='movie_intro_info_r').find_all('span')[-3].text.strip().split('：')[1]
    score=main.find('div',class_="score_num count").text.strip()
    
    if show:
        print(image_url,video_url,imdb,score)
    
    return [image_url,video_url,imdb,score]
```

```python
get_moviedata(datas[-1][-1])
```

```python
movie_data=copy.deepcopy(datas)
for data in movie_data:
    print(f'{data[0]}.{data[1]}抓取中.....')
    data.extend(get_moviedata(data[-1]))
    
print('資料抓取完畢!')    
```

## 儲存圖檔

```python
def save_pic(url,file_name='temp.jpg'):
    try:
        response = requests.get(url)
        with open(file_name, "wb") as f:
            f.write(response.content)
                    
        print('file save.')            
    except Exception as e:
        print(e)        
   
```
