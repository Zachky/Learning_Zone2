> Install selenium

> Import resource

```python
from selenium import webdriver as WD
from selenium.webdriver.chrome.service import Service as Chrome_Sev
from selenium.webdriver.common.by import By
import Zach_Library as Zach
```

> Use Service & Webdriver

```python
strPath = 'C:\Web driver\Chrome\chromedriver.exe'

ServiceChrome = Chrome_Sev(strPath)
ServiceChrome
```

```python
url = 'https://www.ibon.com.tw/retail_inquiry.aspx#gsc.tab=0'
Test_Chrome = WD.Chrome(service=ServiceChrome)
Test_Chrome.get(url)
```

```python
Test_Chrome.page_source
```

```python
Soup = Zach.BeautifulSoup(Test_Chrome.page_source,'lxml')
Soup
```

```python
trs = Soup.find(id='InquiryResule').find('table').find_all('tr')
trs
```

```python
for tr in trs:
    for th in tr.find_all('th'):
        print(th.text.strip(),end='\t')
    
    for td in tr.find_all('td'):
        print(td.text.strip(),end='\t')

    print()
```

> Xpath for element

> Get InputArea Elemental

```python
xpath = '/html/body/form/div[4]/div/div/div/div[1]/div[2]/table/tbody/tr[3]/td/input'
InputZone = Test_Chrome.find_element(By.XPATH,xpath)
InputZone
```

```python
InputZone.clear()
InputZone.send_keys('文山區')
```

> Get QueryButton Elemental

```python
buttonPath = '/html/body/form/div[4]/div/div/div/div[1]/div[2]/table/tbody/tr[3]/td/a'
QButton = Test_Chrome.find_element(By.XPATH,buttonPath)
QButton.click()
```

```python
Test_Chrome.quit()
```
