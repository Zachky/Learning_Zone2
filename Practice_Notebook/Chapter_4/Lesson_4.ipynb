```python
api_url = "https://bendon-dao.com/shops/datas?"
```

```python
form_data={
    'county':'新北市',
    'district':'all',
    'shop_category_id':'all',
    'keyword':'蛋糕',
    'from':0,
    'limit':5000
}
```

```python
Resp = req.get(api_url,form_data)
JS_Result = Resp.json()
JS_Result['shops']
```

```python
len(JS_Result['shops'])
```

```python
JS_Result['shops'][0]
```

```python
JS_Result['shops'][0]['categories'][0]['name']
```

```python
ColumnList = ['商家名稱','類別','地址']
```

```python
DataRow = []
DataSet = []

Name = ''
Address = ''
Category = ''

for item in JS_Result['shops']:
    
    Name = item['shop_name']
    Address = f"{item['county']}{item['district']}{item['address']}"
    # or Address = ''.join(item['county'],item['district'],item['address'])
    Category = item['categories'][0]['name']
    
    DataRow.append(Name)
    DataRow.append(Category)
    DataRow.append(Address)

    DataSet.append(DataRow)
    DataRow = []

DataSet
```

```python
df = pd.DataFrame(DataSet,columns=ColumnList)
df
```

```python
today = dt.now().strftime("%Y%m%d_%H%M%S")
df.to_csv(f'./Output/BendonShopList_{today}.csv',encoding='utf-8-sig')
```
