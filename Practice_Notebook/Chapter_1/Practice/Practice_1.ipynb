```python
import requests as req
from bs4 import BeautifulSoup as BSoup
import webbrowser as wb
import pandas as pd

url="https://tw.yahoo.com"
Response = req.get(url)
Response.encoding='utf-8'
SoupContent = BSoup(Response.text, 'lxml');SoupContent
```

```python
SoupContent.find_all('b',class_='Hidden')
```

```python
SoupContent.find_all('b',string='youtube')
```

### Find all a tag which has 'attrs' attribute, export as dictionary which contain link and its text

```python
[ {i.text:i['href']} for i in SoupContent.find_all('a',attrs={'data-ylk': True }) ]  
```

### Find all a tag which has 'attrs' attribute, make a dataset then export as excel file

```python
datas = [ [i.text,i['href']] for i in SoupContent.find_all('a',attrs={'data-ylk': True }) ]  
datas
```

```python
dataset = pd.DataFrame(datas, columns=['Name','Link'])
dataset
```

```python
dataset.to_csv('yahoo_link.csv',encoding='utf-8-sig', index=0)
```
