# Request Mode Practice

### 1. Get response code

```python
import requests

url = 'https://tw.yahoo.com/'
strResponse=requests.get(url)
print(strResponse)
```

> Response code table:
[HTTP response status codes](https://developer.mozilla.org/en-US/docs/Web/HTTP/Status)

### 2. Get Response Content

```python
strResponse.text
```

```python
requests.codes.ok
```

```python
strResponse.status_code==requests.codes.ok
```

### 3. Get content with Encoding (utf-8)

```python
strResponse = requests.get(url)
strResponse.encoding = 'utf-8'

if strResponse.status_code == 200:
    print(strResponse.text)
```

```python
type(strResponse.text)
```

### 5. Find Function Example

##### * Find the first obj of the given elemental

```python
SoupContent.find('title')
```

```python
SoupContent.find('title').text
```

##### * Get the value of given attribute of given elemental

```python
SoupContent.find('a')['href']
```

```python
SoupContent.find('a').get('id')
```

##### * Find all obj of given elemental

```python
SoupContent.find_all('a')
```

```python
type(SoupContent.find_all('a'))
```

##### * Practice: Get link name and link address

```python
for obj in SoupContent.find_all('a'):
  print(obj.text)
  print(obj['href'])
```

##### * Advanced Practice: List Comprehension (Export as 2D List)

```python
[[ link.text,link['href'] ] for link in SoupContent.find_all('a')]
```

##### * Advanced Practice: List Comprehension (Export as Dictionary)

```python
{ link.text:link['href'] for link in SoupContent.find_all('a') }
```

##### * Find different tag simultaneously

```python
SoupContent.find_all(['title','a'])
```

##### * Find all elemental by using attribute 

```python
List_result = SoupContent.find_all(class_='Hidden');List_result
```

##### Practice : Formate the output

```python
for index in range(0,len(List_result)):
  print(List_result[index].text)
```

> String.join require a list of obj as parameter, hence the for loop need to traversal each elemental in "List_result"

```python
print('\n'.join( content.text for content in List_result ) )
```
