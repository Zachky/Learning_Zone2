### 產生天氣預報圖

```python
def get_temperature(low,high,days=30):
    temp=np.random.uniform(low,high,days).round(2)
    return temp
```

```python
tokyo=get_temperature(-10,32)
taipei=get_temperature(10,35)
bangkok=get_temperature(28,40)

tokyo,taipei,bangkok
```

### 加上figure底板，grid，新增label，legend，title

```python
plt.figure(figsize=(10,3))
plt.grid(True)

x = pd.date_range("2024-01-01",periods=30).strftime("%Y-%m-%d")
plt.plot(x,tokyo,"s:m",label="tokyo")
plt.plot(x,taipei,"s:m",label="taipei")
plt.plot(x,bangkok,"s:m",label="bangkok")

plt.legend(loc=0)

plt.xticks(rotation=90)

plt.title("weather report",fontsize=20, pad=20)
plt.xlabel("day", fontsize=12, labelpad=20)
plt.ylabel("temperature", fontsize=12, labelpad=40)

plt.show()
```

```python
df=pd.read_csv("統一營收表_202007.csv",thousands=",")[::-1]
df
```

```python
df.info()
```

```python
dates = df[df['當月營收']<df['去年同月營收']].index
date_index = []

for date in dates:
    for i in range(len(df.index)):
        if date == df.index[i]:
            date_index.append(i)

date_index
```

```python
plt.figure(figsize=(20,8))
plt.grid(True)

styles=["^--b","o-.c","^:m"]
column=["Tokyo", "Taipei", "Avg Temp"]

plt.plot(df['月份'],df['當月營收']/10000,"^--b",label="當月營收")
plt.plot(df['月份'],df['去年同月營收']/10000,"o-.c",label="去年同月營收")

plt.legend(loc='upper left', prop={'size': 30} , bbox_to_anchor=(0,1))

plt.xticks(rotation=90)

for index in date_index:
    date=df.index[index]
    plt.text(index+0.05,df.loc[date]["當月營收"]/10000,date,color="black",rotation=90)
    plt.text(index,df.loc[date]["當月營收"]/10000,date,color="red",rotation=90)

plt.title("統一營收表",fontsize=20, pad=20)
plt.xlabel("月份", fontsize=12, labelpad=20)
plt.ylabel("金額", fontsize=12, labelpad=40)

plt.show()
```
