> Scrap NBA Scores, use 2024/02-15 as example, get scores from each game

```python
url = "https://tw-nba.udn.com/nba/schedule_boxscore/2024-02-15"
soup = getSoup(url)
soup
```

```python
games=soup.find('div',class_='cards').find_all('div',class_='card')
games
```

```python
link = games[0].find('a')['href']
link
```

```python
soup = getSoup(link)
soup
```

> Get Scores

```python
for tr in soup.find('table',class_='stable boxscore').find_all('tr'):
    for th in tr.find_all('th'):
        print(th.text.strip(),end='\t')
    
    for td in tr.find_all('td'):
        print(td.text.strip(),end='\t')

    print()

    
```
