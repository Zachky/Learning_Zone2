```python
def get_car_data(url):

    datas=[]
    chrome=None
    try:
        chrome=Zach.get_chrome(url,hide=True)
        soup=Zach.BeautifulSoup(chrome.page_source,'lxml')
        cars=soup.find(id='result-items').find_all('a')

        for i,car in enumerate(cars):
            try:
                title=car.find('span',class_="ib-it-text").text.strip()
                link=url+car.get('href')
                Image_url=car.find('img').get('data-src')
                location=car.find('span',class_="ib-ii-item").text.strip()
                price=car.find('span',class_="ib-price").text.split()[0]
                years,km=[ span.text.strip() for span in car.find_all('span',class_="ib-icon") ]
                datas.append([title,link,Image_url,location,price,years,km])

            except Exception as e:
                print(e)

    except Exception as e:
        print(e)
    finally:
        if chrome is not None:
            chrome.quit()

    return datas
```

```python
def Price_Convert(price):
    new_price = eval(price.replace('萬',''))
    if '萬' in price:
        new_price = eval(price.replace('萬',''))*10000
    return new_price
```

```python
df1['price'] = df1['price'].apply(Price_Convert)
```
