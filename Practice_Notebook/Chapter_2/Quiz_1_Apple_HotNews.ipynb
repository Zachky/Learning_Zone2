# Web Scraping Practice

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

### Define GetSoup function for future use

```python
def getSoup(url):
  try:
    resp=requests.get(url)
    resp.encoding='utf-8'
    if resp.status_code == 200:
      return BeautifulSoup(resp.text,'lxml')
    else:
      return 'Error:Status_code'+ str(resp.status_code)

  except Exception as e:
    return e
```

### Apple News - Hot News

```python
url = 'https://tw.nextapple.com/realtime/latest'
SoupContent = getSoup(url);SoupContent
```

##### 1. Find the list of hot news title

> Bigger Area

```python
lis=SoupContent.find_all('div')
lis
```

>Correct Area

```python
lis = SoupContent.find('div',class_='block-inner').find('ul').find_all('li')
lis
```

> Length of datas

```python
len(lis)
```

> Extract data from web page as list

```python
for item in lis:
  print(f'( {item.text.replace('\n','')} , {item.find('a')['href']} )')
```

##### 2. Export data as csv

```python
datas = [[ item.text.replace('\n','') , item.find('a')['href']] for item in lis ]
datas
```

```python
df = pd.DataFrame(datas, columns=['Title','Link'])
df

df.to_csv('./Output/Quiz_1/Apple_HotNews.csv',encoding='utf-8-sig')
```
