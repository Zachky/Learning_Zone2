```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("./weather.csv",index_col=0)
df
```

```python
plt.figure(figsize=(20,8))

plt.bar(df.index,df["tokyo"],label="tokyo",edgecolor="red", lw=0.5)
plt.bar(df.index,df["taipei"],alpha=0.5,label="taipei",width=0.8)

#Avg Temp
plt.plot(df.mean(axis=1),"o-c",lw=2)

plt.xticks(rotation=90)

plt.legend()

plt.show()
```
