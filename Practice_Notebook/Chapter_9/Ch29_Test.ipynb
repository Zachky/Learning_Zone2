```python
def is_Data_exists(mday,sno):
    Result = False
    sqlstr =f"select * from taipei where mday='{mday}' and sno={sno};"
    datas = list(cursor.execute(sqlstr))
    if len(datas) > 0:
        Result = True

    return Result
```

```python
import os

db_name="ubike.db"
conn,cursor=None,None
def get_ubike_data_tosql():
    global conn,cursor
    try:
        
        df=pd.read_json(api_url)[["mday","sno","sarea","sna","tot","sbi","bemp","lat","lng","ar","act"]]
        df=df[ df["act"]==1 ]

        if not os.path.exists(db_name):
            conn=sqlite3.connect(db_name)
            df.to_sql("taipei",con=conn,if_exists="append",index=None)
            print("Database initialize completed!")
        else:
            conn=sqlite3.connect(db_name)
            print("Database existed!")
        # conn=sqlite3.connect("ubike.db")
        # cursor=conn.cursor()
        # print('db open.')
        
    except Exception as e:
        print(e)
    finally:
        if conn is not None:
            conn.close()
            print('db close.')
```
